# Loading the model with 4-bit quantization and with prompt lookup decoding


```python
from vllm import LLM
import torch
model_id = "mistralai/Mistral-7B-v0.1"
llm = LLM(
    model=model_id,
    tokenizer = model_id,
    seed= 42,
    tensor_parallel_size = 1,
    dtype=torch.float16,
    trust_remote_code=True,
    quantization="bitsandbytes",
    load_format="bitsandbytes",
    gpu_memory_utilization=0.5, #increase this based on availaible memory
    max_model_len=1024,
    enforce_eager=False,
    speculative_model="[ngram]",
    num_speculative_tokens=1,
    ngram_prompt_lookup_max=10,  #Tuning this for better performance
    use_v2_block_manager=True,
    # disable_sliding_window=True,
    # enable_chunked_prefill=True
)
```

```python
import time
from vllm import LLM, SamplingParams

def warmup_model(llm, warmup_prompt="Warmup prompt", sampling_params=None):
    """
    Run a warmup inference to get the model ready and avoid initial cold start latencies.
    :param llm: The language model instance (e.g., vLLM instance).
    :param warmup_prompt: A dummy prompt to warm up the model.
    :param sampling_params: The sampling parameters to use for the generation (can be default or custom).
    """
    if sampling_params is None:
        # Default sampling parameters if none are provided
        sampling_params = SamplingParams(
            temperature=0.8,
            top_p=0.95,
            top_k=6,
            max_tokens=10,  # Short token count for fast warmup
            min_tokens=5
        )

    print("Warming up the model...")
    start_time = time.time()

    # Perform a quick generation with a small prompt
    _ = llm.generate([warmup_prompt], sampling_params)

    end_time = time.time()
    print(f"Model warmup completed in {end_time - start_time:.4f} seconds.")


warmup_model(llm)
```

# FOR SINGLE USER INPUT AND PERFORMANCE DISPLAY

```python
import time


def get_user_input():
    """Function to take user input for the prompt."""
    return input("Enter your prompt: ")

def calculate_throughput(num_tokens, time_taken):
    """Calculate throughput in tokens per second."""
    return num_tokens / time_taken if time_taken > 0 else 0

def main():
    # Initialize sampling parameters
    sampling_params = SamplingParams(
        temperature=0.8,
        top_p=0.95,
        top_k=3,
        max_tokens=30,
        min_tokens=30
    )

    # Step 1: Get user prompt
    user_prompt = get_user_input()


    # Combine the system prompt with the user prompt
    full_prompt = '{"role": "system-chatbot", "content": "' + user_prompt + '"}'

    # Step 2: Measure the time taken for generation
    start_time = time.time()

    # Step 3: Pass the formatted prompt to the model
    outputs = llm.generate([full_prompt], sampling_params)

    # Step 4: Measure the end time
    end_time = time.time()
    total_time = end_time - start_time

    # Step 5: Extract tokens and generated text
    output = outputs[0]  # Assuming only one output
    generated_text = output.outputs[0].text
    input_tokens = len(output.prompt.split())  # Token count for input
    output_tokens = len(generated_text.split())  # Token count for output
    total_tokens = input_tokens + output_tokens

    # Step 6: Calculate throughput
    throughput = calculate_throughput(total_tokens, total_time)

    # Step 7: Print results
    print(f"\nPrompt: {full_prompt!r}")
    print(f"Generated text: {generated_text!r}")
    print(f"Input tokens: {input_tokens}, Output tokens: {output_tokens}")
    print(f"Total tokens: {total_tokens}")
    print(f"Time taken: {total_time:.4f} seconds")
    print(f"Throughput: {throughput:.2f} tokens/second")

if __name__ == "__main__":
    main()


```
